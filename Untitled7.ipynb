```python
import os
import pandas as pd
import numpy as np

import kagglehub

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt

path = kagglehub.dataset_download("saurabhshahane/ecommerce-text-classification")
print("Папка с датасетом:", path)
print("Файлы:", os.listdir(path))
csv_files = [file for file in os.listdir(path) if file.endswith(".csv")]
print("CSV файлы:", csv_files)
csv_path = os.path.join(path, csv_files[0])
df = pd.read_csv(csv_path, header=None)
df.columns = ["category", "text"]
print(df.head())
print(df.info())
print(df["category"].value_counts())
df = df.dropna()
X = df["text"]
y = df["category"]
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
print("Классы:", label_encoder.classes_)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

model = Pipeline([
    ("tfidf", TfidfVectorizer(
        stop_words="english",
        max_features=50000,
        ngram_range=(1, 2)
    )),
    ("svd", TruncatedSVD(
        n_components=300,
        random_state=42
    )),
    ("classifier", LogisticRegression(
        max_iter=1000,
        random_state=42
    ))
])



model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(
    y_test,
    y_pred,
    target_names=label_encoder.classes_
))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(range(len(label_encoder.classes_)), label_encoder.classes_, rotation=45)
plt.yticks(range(len(label_encoder.classes_)), label_encoder.classes_)

for i in range(len(cm)):
    for j in range(len(cm[i])):
        plt.text(j, i, cm[i][j], ha="center", va="center")

plt.tight_layout()
plt.show()
test_texts = [
    "Wireless bluetooth headphones with noise cancellation",
    "Cotton t-shirt for men with round neck",
    "A book about machine learning and artificial intelligence",
    "Stainless steel kitchen knife set for home cooking"
]

test_predictions = model.predict(test_texts)
test_predictions_labels = label_encoder.inverse_transform(test_predictions)

for text, prediction in zip(test_texts, test_predictions_labels):
    print("Текст:", text)
    print("Класс:", prediction)
    print()
```
